```python
def extract_message(lines, prefix=):
    words = set()
    is_message = False
    for ln in lines:
        if is_message:
            words = words.union({i for i in ln.lower().split() if i.isalpha()})
        
        
        elif ln.startswith(b'X-Spam-Level:'):
            is_message = True
    
    return words
```
